```python
import pandas as pd
```

```python
df = pd.read_csv("Hotel Reservations.csv")
```

```python
df = df.drop(["Booking_ID", "arrival_year"], axis=1)
df["booking_status"] = df["booking_status"].apply(lambda x : 1 if x == "Canceled" else 0)
```

```python
df = pd.get_dummies(df, dtype=int)
```
